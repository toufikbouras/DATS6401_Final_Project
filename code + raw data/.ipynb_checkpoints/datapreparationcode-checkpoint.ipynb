```python
# Importing the unemployement dat, processing it and joining all the data into one dataframe 
from xlrd import open_workbook 
from os import listdir
import pandas as pd

files = listdir('../code + raw data')
```

```python
# Get the file names 
sheets = [filename for filename in files if filename.endswith("xlsx")]
```

```python
# Print the file names without the extension
for name in sheets:
    print(name.replace('.xlsx',''))
```

```python
# Create a new files contains all states data
dataframe = []
for name in sheets:
    data = pd.read_excel(name, skiprows=10,nrows=18, ncols=13).fillna(method='ffill', axis=1)
    data['State'] = name.replace('.xlsx','')
    dataframe.append(data)
    data = pd.concat(dataframe)    
    

    
```

```python
data
```

```python
# Create the average column
data['Mean'] = round(data[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec']].mean(axis=1), 2)
```

```python
# Save the data to a csv file 
data.to_csv('unempolyemnt data all states.csv')
```
